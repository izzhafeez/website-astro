```python
import os
import json
```

```python
input_path = "../../public/data/name"
slugs = []
for o in os.listdir(input_path):
    if not o.endswith(".json"):
        continue
    file_slug = o[:-5]
    slugs.append(file_slug)

with open(os.path.join("../../src/data", "name-slugs.json"), "w", encoding="utf-8") as f:
    json.dump(slugs, f, ensure_ascii=False, indent=4)
```
